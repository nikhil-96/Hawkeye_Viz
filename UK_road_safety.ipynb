##### Vehicles 2016-2020

```python
dfv = pd.read_csv('https://data.dft.gov.uk/road-accidents-safety-data/dft-road-casualty-statistics-vehicle-last-5-years.csv')
print('Records:', dfv.shape[0], '\nColumns:', dfv.shape[1])
dfv.columns = dfv.columns.str.title()
dfv.head()
```

Next, let's define a new column that groups the Time the accidents happened into one of five options:

* Morning
* Afternoon
* Rush_Hour
* Evening
* Night

```python
# slice first and second string from time column
dfa['Hour'] = dfa['Time'].str[0:2]

# convert new column to numeric datetype
dfa['Hour'] = pd.to_numeric(dfa['Hour'])

# drop null values in our new column
dfa = dfa.dropna(subset=['Hour'])

# cast to integer values
dfa['Hour'] = dfa['Hour'].astype('int')
```

```python
# define a function that turns the hours into daytime groups
def when_was_it(hour):
    if hour >= 5 and hour < 10:
        return "morning rush (5-10)"
    elif hour >= 10 and hour < 15:
        return "office hours (10-15)"
    elif hour >= 15 and hour < 19:
        return "afternoon rush (15-19)"
    elif hour >= 19 and hour < 23:
        return "evening (19-23)"
    else:
        return "night (23-5)"
```

```python
# apply thus function to our temporary hour column
dfa['Daytime'] = dfa['Hour'].apply(when_was_it)
dfa[['Time', 'Hour', 'Daytime']].head(8)
```
